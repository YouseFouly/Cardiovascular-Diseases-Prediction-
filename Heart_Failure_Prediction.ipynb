**Importing Dependincies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn import metrics
```

**Importing Data**

```python
heart_data = pd.read_csv('/content/heart failier.csv')
```

**Label Encoding**

```python
heart_data['Sex'] = heart_data['Sex'].apply(lambda x:1 if x=='M' else 0)
heart_data['ChestPainType'] = heart_data['ChestPainType'].apply(lambda x:0 if x=='ATA' else (1 if x=='NAP' else (2 if x=='ASY' else 3)))
heart_data['RestingECG'] = heart_data['RestingECG'].apply(lambda x:0 if x=='Normal' else (1 if x=='ST' else 2))
heart_data['ExerciseAngina'] = heart_data['ExerciseAngina'].apply(lambda x:1 if x=='Y' else 0)
heart_data['ST_Slope'] = heart_data['ST_Slope'].apply(lambda x:0 if x=='Up' else (1 if x=='Flat' else 2))

```

```python
heart_data.head()
```

```python
heart_data.describe()
```

```python
x = heart_data.drop(columns='HeartDisease',axis=1)
y = heart_data['HeartDisease']
```

```python
print (x)
print (y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state = 2)
```

```python
print (x.shape, x_train.shape, x_test.shape)
```

**Model Training**

```python
model = XGBClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
#Accuracy for prediction on training data
training_data_prediction = model.predict(x_train)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
